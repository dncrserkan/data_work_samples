**&emsp;&emsp;!Values should be less than one because they represent percentages**

**&emsp;&emsp;Create mask and see if any of the columns have values above 1**

```python
mask = (df.iloc[:,1:] > 1)    # do not look at date ('Tarih') column
df[mask.any(axis=1)]
```

<p style="background-image: linear-gradient(#0D9276, #FFFFFF 10%)"> &nbsp </p>

**&emsp;&emsp;But, we also see the values very close to 1. First only look at the approximate value of 1**

```python
mask = (df.iloc[:,1:] > 1) & (df.iloc[:,1:] < 1.001)
df[mask.any(axis=1)]
```

**&emsp;&emsp;971 rows has values smaller than 1.001 and they should remain as they are**

<p style="background-image: linear-gradient(#0D9276, #FFFFFF 10%)"> &nbsp </p>

**&emsp;&emsp;Update the mask**

```python
mask = df.iloc[:,1:] > 1.001
df[mask.any(axis=1)]
```

**&emsp;&emsp;! These values are shown as percentages, these values must be divided by 100 to place them between 0 and 1**
