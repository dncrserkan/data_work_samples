```python
import pandas as pd
import numpy as np

from mlxtend.frequent_patterns import apriori, association_rules
from pyECLAT import ECLAT
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

```python
data = pd.read_csv('data/online_retail_II_clean.csv')
data.head(3)
```

```python
data_UK = data[data['Country']=='United Kingdom'][['Invoice', 'Description', 'Quantity']]
data_UK.head(3)
```

```python
print(data_UK['Description'].str.contains('&').sum())
data_UK['Description'] = data_UK['Description'].str.replace('&', 'AND')
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

```python
data_UK = data_UK.groupby(['Invoice', 'Description']).sum()
data_UK = data_UK.unstack(fill_value=0)

data_UK.columns = data_UK.columns.levels[1]
data_UK.reset_index(drop=True, inplace=True)
data_UK.sample(3)
```

```python
# data_UK = data_UK.map(lambda x: 0 if x==0 else 1)
data_UK = data_UK.map(lambda x: False if x==0 else True)
data_UK.sample(3)
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

```python
df_1 = apriori(data_UK, min_support=0.02, use_colnames=True)
# display(df_1)

df_2 = association_rules(df_1, metric='confidence') # min_threshold=0.7
display(df_2)

df_3 = association_rules(df_1, metric='lift', min_threshold=1)
display(df_3.nlargest(5, columns='lift'))
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

```python
columns = df_1.explode(column='itemsets')
columns['itemsets'].str.split(',')
print(data_UK.shape[1], columns.shape[0], columns['itemsets'].nunique())
columns = columns['itemsets'].unique()
```

```python
df = data_UK[[*columns]].copy()
df.tail(3)
```

```python
df = df.apply(lambda col: [col.name if val else np.nan for val in col])
df.columns = np.arange(df.shape[1])
display(df.tail(3))
```

```python
eclat = ECLAT(df, verbose=True)
eclat.df_bin.tail(3)
```

```python
_1 = min_product = 1
_2 = max_product = 2     # eclat.df_bin.sum(axis=1).max() # DO NOT TRY
_3 = min_support = 0.02

a, b = eclat.fit(min_combination=_1, max_combination=_2, min_support=_3, verbose=True)
display(pd.Series(b).to_frame())
```

```python
df_4 = pd.DataFrame(b.items())
df_4.rename(columns={0:'itemsets', 1:'support'}, inplace=True)
df_4['itemsets'] = df_4['itemsets'].str.split(' & ')
df_4['itemsets'] = df_4['itemsets'].map(frozenset)
# display(df_4)
```

```python
df_5 = association_rules(df_4, metric='confidence')
display(df_5)

df_6 = association_rules(df_4, metric='lift', min_threshold=1)
display(df_6.nlargest(5, columns='lift'))
```
