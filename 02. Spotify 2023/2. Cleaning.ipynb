```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('data/spotify-2023.csv', encoding='latin-1')
df.head()
```

<p style="background-image: linear-gradient(#0D9276, #FFFFFF 10%)"> &nbsp </p>

#### Drop unnecessary columns (for me)

```python
single1 = df[['in_spotify_playlists']]
single2 = df[['in_spotify_charts']]
range1 = df.loc[:,'in_apple_playlists':'in_shazam_charts']
filt = single1 + single2 + range1
# df.drop(columns=filt.columns)    # Always check first
df.drop(columns=filt.columns, inplace=True) 
df.tail()    # Just for see another range of data
```

<p style="background-image: linear-gradient(#0D9276, #FFFFFF 10%)"> &nbsp </p>

#### Create date column and drop old ones

```python
temp_df = pd.DataFrame()
temp_df['year'] = df['released_year'].astype(str)
temp_df['month'] = df ['released_month'].astype(str).str.zfill(2)
temp_df['day'] = df['released_day'].astype(str).str.zfill(2)

df['released_date'] = pd.to_datetime(temp_df[['year', 'month', 'day']])
del(temp_df)
filt = df.loc[:,'released_year':'released_day']
# df.drop(columns=filt.columns)   # Always check first
df.drop(columns=filt.columns, inplace=True)
df.head()
```

<p style="background-image: linear-gradient(#0D9276, #FFFFFF 10%)"> &nbsp </p>

#### Change names of columns | del '_%' if exist

```python
# df.columns
for col in df.columns:
    if '_%' in col:
        new = col[:-2]
        df.rename(columns={col:new}, inplace=True)
df.head()
```

<p style="background-image: linear-gradient(#0D9276, #FFFFFF 10%)"> &nbsp </p>

#### Separate artist(s)_name column and drop it

```python
temp_df = pd.DataFrame()
temp_df = df['artist(s)_name'].str.split(', ', expand=True)
feat_df = pd.DataFrame(temp_df.iloc[:,1:])

df['artist'] = temp_df[[0]]
df['featuring_artists'] = feat_df.iloc[:,0:7].apply(lambda x: ', '.join(filter(None, x)), axis=1)
df['featuring_artists'].replace({pd.NA: None, None:None, '': None}, inplace=True)
df.drop(columns='artist(s)_name', inplace=True)
del(temp_df, feat_df)
df.tail()
```

<p style="background-image: linear-gradient(#0D9276, #FFFFFF 10%)"> &nbsp </p>

#### Arrange columns order

```python
new_order =['track_name', 'artist', 'featuring_artists','artist_count', 'streams', 'bpm',
            'key', 'mode', 'danceability', 'valence', 'energy', 'acousticness',
            'instrumentalness', 'liveness', 'speechiness', 'released_date']
df = df.loc[:,new_order]
df.head(3)
```

<p style="background-image: linear-gradient(#0D9276, #FFFFFF 10%)"> &nbsp </p>

#### Change dtypes of 'streams' column

```python
df['streams'].info()
# df['streams'] = df['streams'].astype(int)    # ERROR
```

**Fixing streams column**<br>
    → find non-numeric value

```python
df['streams'].max()
# real value on Spotify → 224.800.316
```

```python
filt = df['streams'] == df['streams'].max()
# df[filt]['streams'] = '224800316'    # ERROR
df.loc[filt, 'streams'] = '224800316'
df[filt]
```

**Now ready to change**

```python
display(df['streams'].dtypes)
df['streams'] = df['streams'].astype(int)
display(df['streams'].dtypes)
```

<p style="background-image: linear-gradient(#0D9276, #FFFFFF 10%)"> &nbsp </p>

**Fix some values**<br>
    → min streams values (first two) are too small for this dataset
    → find min value (2 times)

```python
df.sort_values(by='streams').head()    # first see min values
# real value on Spotify -> 557.890.134 and 179.111.799
```

```python
correct_values = [557890134, 179111799]    # order is important first the minimum than one bigger from data
for value in correct_values:
    filt = df['streams'] == df['streams'].min()
    df.loc[filt, 'streams'] = value

df.iloc[[123, 393],]    # see new values
```
